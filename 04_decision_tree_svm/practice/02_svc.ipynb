# 실습 2. SVC 분류

Breast Cancer 데이터와 Moons 데이터를 사용해 SVC의 핵심 흐름을 연습함.

- `StandardScaler`로 feature 스케일을 맞춤.
- 선형 SVC를 학습하고 accuracy, classification report를 확인함.
- `C` 값 변화에 따라 score와 support vector 개수가 어떻게 달라지는지 비교함.
- RBF 커널에서 `gamma` 값 변화에 따른 성능 차이를 확인함.


````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_breast_cancer, make_moons
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

cancer = load_breast_cancer(as_frame=True)
cancer_df = cancer.frame.copy()

cancer_X = cancer_df[cancer.feature_names]
cancer_y = cancer_df['target']

print(cancer_df.shape)
print(cancer.target_names)
display(cancer_df.head())
````

## 문제 1. 학습/평가 데이터 분리와 스케일링

`cancer_X`, `cancer_y`를 학습/평가 데이터로 나누고 `StandardScaler`를 적용함.

힌트:
- `train_test_split()`
- `stratify=cancer_y`
- `scaler.fit_transform(X_train)`
- `scaler.transform(X_test)`

### 실행 결과
```text
X_train: (455, 30)
X_test: (114, 30)
y_train: (455,)
y_test: (114,)
```


````python
X_train, X_test, y_train, y_test = train_test_split(
    cancer_X,
    cancer_y,
    test_size=0.2,
    random_state=42,
    stratify=cancer_y
)
scaler = StandardScaler()

X_train_t = scaler.fit_transform(X_train)
X_test_t = scaler.transform(X_test)
print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)

````

## 문제 2. 선형 SVC 모델 학습과 평가

`kernel='linear'`인 SVC 모델을 만들고 학습셋/평가셋 accuracy와 classification report를 확인함.

힌트:
- `SVC(kernel='linear', C=1.0)`
- `fit()`
- `score()`
- `classification_report()`

### 실행 결과
```text
학습셋 accuracy: 0.9912087912087912
평가셋 accuracy: 0.9736842105263158

              precision    recall  f1-score   support

   malignant       0.95      0.98      0.96        42
      benign       0.99      0.97      0.98        72

    accuracy                           0.97       114
   macro avg       0.97      0.97      0.97       114
weighted avg       0.97      0.97      0.97       114
```


````python
svc = SVC(kernel='linear', C=1.0)
svc.fit(X_train_t, y_train)

y_pred = svc.predict(X_test_t)

print("학습셋 accuracy: ", svc.score(X_train_t, y_train))
print("평가셋 accuracy: ", svc.score(X_test_t, y_test))

print(classification_report(y_test, y_pred))
````

## 문제 3. `C` 값 변화에 따른 성능 비교

`C` 후보를 바꿔가며 선형 SVC를 학습하고 학습셋/평가셋 accuracy, support vector 개수를 비교함.

힌트:
- `for c_value in [...]`
- `SVC(kernel='linear', C=c_value)`
- `len(model.support_)`

### 실행 결과
```text
C  train_accuracy  test_accuracy  support_vector_count
0    0.01        0.978022       0.964912                    99
1    0.10        0.984615       0.982456                    51
2    1.00        0.991209       0.973684                    32
3   10.00        0.991209       0.982456                    26
4  100.00        0.995604       0.973684                    26
```


````python
svc_df = pd.DataFrame(columns=['C', 'train_accuracy', 'test_accuracy', 'support_vector_count'])
for c_value in [0.01, 0.1, 1, 10, 100]:
    svc = SVC(kernel='linear', C=c_value)
    svc.fit(X_train_t, y_train)
    svc_df.loc[len(svc_df)] = [c_value, svc.score(X_train_t, y_train), svc.score(X_test_t, y_test), len(svc.support_)]
svc_df
````

## 문제 4. `C` 변화 결과 시각화

문제 3에서 만든 `c_result_df`를 사용해 accuracy와 support vector 개수 변화를 그래프로 확인함.

### 실행 결과
```text
그래프 출력: Accuracy by C, Support vectors by C
```



````python
# 문제 풀이 후 실행: C 변화 결과 시각화

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

sns.lineplot(data=svc_df, x='C', y='train_accuracy', marker='o', label='train', ax=axes[0])
sns.lineplot(data=svc_df, x='C', y='test_accuracy', marker='o', label='test', ax=axes[0])
axes[0].set_xscale('log')
axes[0].set_title('Accuracy by C')
axes[0].set_ylim(0.85, 1.01)

sns.lineplot(data=svc_df, x='C', y='support_vector_count', marker='o', ax=axes[1])
axes[1].set_xscale('log')
axes[1].set_title('Support vectors by C')

plt.tight_layout()
plt.show()
````

## 문제 5. RBF 커널과 `gamma` 값 비교

Moons 데이터는 직선 하나로 나누기 어려운 비선형 데이터임. RBF 커널 SVC로 `gamma` 값을 바꿔가며 성능을 비교함.

힌트:
- `make_moons()`
- `SVC(kernel='rbf', C=1.0, gamma=gamma_value)`
- `StandardScaler`

### 실행 결과
```text
gamma  train_accuracy  test_accuracy  support_vector_count
0    0.1        0.871111       0.840000                    85
1    1.0        0.915556       0.893333                    69
2   10.0        0.960000       0.920000                   132
3   30.0        0.977778       0.920000                   186
```


````python
moon_X, moon_y = make_moons(n_samples=1000, noise=0.2, random_state=42)

moon_X_train, moon_X_test, moon_y_train, moon_y_test = train_test_split(
    moon_X,
    moon_y,
    test_size=0.2,
    random_state=42,
    stratify=moon_y
)

scaler = StandardScaler()
moon_X_train_t = scaler.fit_transform(moon_X_train)
moon_X_test_t = scaler.transform(moon_X_test)

scaler.fit_transform(moon_X_train)
scaler.transform(moon_X_test)

gamma_result_df = pd.DataFrame(columns=['gamma', 'train_accuracy', 'test_accuracy', 'support_vector_count'])
for gamma_value in [0.1, 1.0, 10, 30]:
    svc = SVC(kernel='rbf', C=1.0, gamma=gamma_value)
    svc.fit(moon_X_train_t, moon_y_train)
    gamma_result_df.loc[len(gamma_result_df)] = [gamma_value, svc.score(moon_X_train_t, moon_y_train), svc.score(moon_X_test_t, moon_y_test), len(svc.support_)]

gamma_result_df
````

## 문제 6. `gamma` 변화 결과 시각화

`gamma_result_df`를 사용해 `gamma` 값에 따른 학습셋/평가셋 accuracy 변화를 그래프로 확인함.

### 실행 결과
```text
그래프 출력: RBF SVC accuracy by gamma
```



````python
# 문제 풀이 후 실행: gamma 변화 결과 시각화

plt.figure(figsize=(6, 4))
sns.lineplot(data=gamma_result_df, x='gamma', y='train_accuracy', marker='o', label='train')
sns.lineplot(data=gamma_result_df, x='gamma', y='test_accuracy', marker='o', label='test')
plt.xscale('log')
plt.ylim(0.75, 1.01)
plt.title('RBF SVC accuracy by gamma')
plt.show()
````
